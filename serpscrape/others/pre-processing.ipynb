```python
import csv
import os
import requests

# Step 1: Create a folder for images
os.makedirs("product_images", exist_ok=True)

# Step 2: Read the CSV and download images
with open("tshirt_data.csv", newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for idx, row in enumerate(reader, start=1):
        image_url = row.get("thumbnail") or row.get("thumbnails")

        if image_url:
            try:
                response = requests.get(image_url, timeout=10)
                response.raise_for_status()
                filename = f"product_images/tshirt_{idx}.jpg"

                with open(filename, "wb") as f:
                    f.write(response.content)

                print(f"Downloaded: {filename}")
            except Exception as e:
                print(f"Failed to download image {idx}: {e}")
        else:
            print(f"No image URL for item {idx}")
```

```python
import csv
import os
import requests

TARGET_ROW = 788  # Change this to the desired row number (1-indexed)

os.makedirs("product_images", exist_ok=True)

with open("tshirt_data.csv", newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for idx, row in enumerate(reader, start=1):
        if idx == TARGET_ROW:
            image_url = row.get("thumbnail") or row.get("thumbnails")

            if image_url:
                try:
                    response = requests.get(image_url, timeout=10)
                    response.raise_for_status()
                    filename = f"product_images/tshirt_{idx}.jpg"

                    with open(filename, "wb") as f:
                        f.write(response.content)

                    print(f"Downloaded image from row {idx}: {filename}")
                except Exception as e:
                    print(f"Failed to download image from row {idx}: {e}")
            else:
                print(f"No image URL found in row {idx}")
            break  # Exit after downloading the specific row
```

```python
from serpapi import GoogleSearch
import csv

api_key = api_key # replace with your API key

params = {
    "engine": "google_shopping",
    "q": "men t shirts",
    "api_key": api_key,
    "gl": "in",
    "hl": "en"
}

search = GoogleSearch(params)
results = search.get_dict()

# Extract product list
products = results.get("shopping_results", [])

# Define CSV fields
fields = ["title", "price", "source", "product_link", "thumbnail", "thumbnails"]

# Save to CSV
with open("tshirts_data.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fields)
    writer.writeheader()
    
    for product in products:
        writer.writerow({
            "title": product.get("title"),
            "price": product.get("price"),
            "source": product.get("source"),
            "product_link": product.get("product_link"),
            "thumbnail": product.get("thumbnail"),
            "thumbnails": product.get("thumbnails")[0] if product.get("thumbnails") else None
        })

print("Product data saved to tshirts_data.csv")
```

```python
from serpapi import GoogleSearch
import time
import csv

API_KEY = api_key
query = "trending men's t-shirts"
results_all = []

# 10 results per page, up to 500 -> 50 pages
for start in range(0, 500, 10):
    params = {
        "engine": "google",
        "q": query,
        "location": "India",
        "google_domain": "google.com",
        "hl": "en",
        "gl": "in",
        "start": start,
        "api_key": API_KEY
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    
    organic_results = results.get("organic_results", [])
    if not organic_results:
        break  # No more data available

    results_all.extend(organic_results)
    
    print(f"Fetched {len(organic_results)} results from start={start}")
    
    # Respectful scraping: Wait 2s between requests to avoid rate-limiting
    time.sleep(2)

# Save to CSV
with open("500_results.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=["title", "link", "snippet"])
    writer.writeheader()
    for item in results_all:
        writer.writerow({
            "title": item.get("title"),
            "link": item.get("link"),
            "snippet": item.get("snippet")
        })

print(f"Total results fetched: {len(results_all)}")
```

```python
from serpapi import GoogleSearch
import csv
import time

API_KEY = api_key
query = "Men's t-shirt with different trending patterns"

results_all = []

for start in range(0, 1000, 10):  # 10 results per page
    params = {
        "engine": "google_shopping",
        "q": query,
        "hl": "en",
        "gl": "in",
        "api_key": API_KEY,
        "start": start
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    
    shopping_results = results.get("shopping_results", [])
    if not shopping_results:
        break  # Stop if no more results

    for product in shopping_results:
        results_all.append({
            "title": product.get("title"),
            "price": product.get("price"),
            "source": product.get("source"),
            "product_link": product.get("product_link"),
            "thumbnail": product.get("thumbnail"),
            "thumbnails": product.get("thumbnails")[0] if product.get("thumbnails") else None,
            "rating": product.get("rating"),
            "reviews": product.get("reviews")
        })

    print(f"Fetched {len(shopping_results)} results from start={start}")
    time.sleep(2)  # Be polite to the API

# Save to CSV
with open("shopping_results_4.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=[
        "title", "price", "source", "product_link",
        "thumbnail", "thumbnails", "rating", "reviews"
    ])
    writer.writeheader()
    writer.writerows(results_all)

print(f"Total products fetched: {len(results_all)}")
```

```python
from serpapi import GoogleSearch
import csv
import time

API_KEY = api_key

queries = [
    "Color block t-shirts for youth",
    "T-shirts with pocket design for men",
    "Tie-dye printed t-shirts online",
    "Longline t-shirts for tall men"
    "Moisture wicking gym t-shirts for men",
    "Latest Bollywood celebrity style t-shirts",
    "Custom printed name t-shirts online",
    "Travel themed printed t-shirts for men",
    "Minimalist plain t-shirts for daily wear",
    "Multi-color striped t-shirts for men",
    "Raglan sleeve t-shirts for men",
    "Henley neck t-shirts under 500",
    "T-shirts with motivational quotes",
]

all_results = []

for query in queries:
    print(f" Query: {query}")
    for start in range(0, 30, 10):  # Try only 3 pages max per query for now
        params = {
            "engine": "google_shopping",
            "q": query,
            "hl": "en",
            "gl": "in",
            "api_key": API_KEY,
            "start": start
        }

        search = GoogleSearch(params)
        results = search.get_dict()

        if "error" in results:
            print(f" API Error: {results['error']}")
            break

        shopping_results = results.get("shopping_results", [])
        print(f"   Page {start//10 + 1}: {len(shopping_results)} products found")

        if not shopping_results:
            break

        for product in shopping_results:
            all_results.append({
                "query": query,
                "title": product.get("title"),
                "price": product.get("price"),
                "source": product.get("source"),
                "product_link": product.get("product_link"),
                "thumbnail": product.get("thumbnail"),
                "rating": product.get("rating"),
                "reviews": product.get("reviews")
            })

        time.sleep(1.5)

# Save to CSV if any results
if all_results:
    with open("tshirt_products_1.csv", mode="w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=[
            "query", "title", "price", "source", "product_link",
            "thumbnail", "rating", "reviews"
        ])
        writer.writeheader()
        writer.writerows(all_results)
    print(f" Saved {len(all_results)} products to tshirt_products.csv")
else:
    print(" No data collected. Please check queries, API key, or quota.")
```
